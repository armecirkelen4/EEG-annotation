```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the positional encoding
L = 10000  # maximum sequence length
D = 256  # embedding dimension

n = np.arange(0, 3000)  # positions
i = np.arange(0, D)  # embedding dimensions
N, I = np.meshgrid(n, i, indexing='ij')

# Positional encoding formula
def positional_encoding(n, i, L, D):
    if i % 2 == 0:
        return np.sin(n / L ** (i / D))
    else:
        return np.cos(n / L ** ((i - 1) / D))

# Apply positional encoding
encoding = np.zeros((len(n), len(i)))
for ni in range(len(n)):
    for ii in range(len(i)):
        encoding[ni, ii] = positional_encoding(n[ni], i[ii], L, D)

# Plot the heatmap
plt.figure(figsize=(10, 6))
plt.imshow(encoding, aspect='auto', cmap='Purples', extent=[0, D, 0, len(n)])
plt.colorbar()
plt.xlabel('embedding dimension')
plt.ylabel('position')
plt.title('Positional Encoding Heatmap')
plt.show()





```
